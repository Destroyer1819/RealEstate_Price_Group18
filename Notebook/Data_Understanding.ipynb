```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## 1. Load Dataset

```python
df = pd.read_csv('../data/melb_data.csv')
df.head()
```

## 2. Dataset Shape and Info

```python
print('Shape:', df.shape)
print()
df.info()
```

## 3. Statistical Summary

```python
df.describe().T
```

## 4. Missing Values

```python
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0])
```

## 5. Price Distribution

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Price'].dropna(), bins=50, color='steelblue', edgecolor='white')
plt.title('Distribution of House Prices (AUD)')
plt.xlabel('Price (AUD)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print(f"Mean Price: ${df['Price'].mean():,.0f}")
print(f"Median Price: ${df['Price'].median():,.0f}")
print(f"Min Price: ${df['Price'].min():,.0f}")
print(f"Max Price: ${df['Price'].max():,.0f}")
```

## 6. Property Type Distribution

```python
plt.figure(figsize=(8, 5))
df['Type'].value_counts().plot(kind='bar', color='steelblue', edgecolor='white')
plt.title('Property Type Distribution')
plt.xlabel('Type (h=House, u=Unit, t=Townhouse)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 7. Price by Region

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Regionname', y='Price', data=df)
plt.title('House Price by Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 8. Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## 9. Price vs Rooms

```python
sns.boxplot(x='Rooms', y='Price', data=df)
plt.title('Price vs Number of Rooms')
plt.tight_layout()
plt.show()
```

## 10. Price vs Distance from CBD

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Distance'], df['Price'], alpha=0.3, color='steelblue')
plt.title('Price vs Distance from CBD')
plt.xlabel('Distance (km)')
plt.ylabel('Price (AUD)')
plt.tight_layout()
plt.show()
```
